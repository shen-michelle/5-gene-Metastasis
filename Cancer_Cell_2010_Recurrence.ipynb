```python
# call pandas data analysis package
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
import io

df = pd.read_csv("BS Taylor 152.csv")
# Read the file that is uploaded in the directory called "File_name.csv"

df = df.dropna(axis=0, how='all')
# deletes the empty rows

tally = 0

for i in range(1,150):
  j= 4 + i
  data = df.iloc[:, [1,2,j]]
  gene = data.columns[2]
  data = data.replace(to_replace='NP', value=np.NaN)
  data = data.dropna(axis=0, how='any')
  # drop the NP values to cleanup data

  data[gene] = data[gene].astype(float)
  print(gene)
  data = data.sort_values(by=[gene], ascending= True)
  # sort data by gene mRNA value

  size = data[gene].size

  nonrecurrent = pd.DataFrame(data=None, index=None, columns=['Value'])
  recurrent = pd.DataFrame(data=None, index=None, columns=['Value'])
  index = 0

  for m in data["Disease Free Status"]:
    if "DiseaseFree" in m:
      mRNA = data.iloc[index, 2]
      nonrecurrent.loc[len(nonrecurrent)] = [mRNA]
    else:
      if "Recurred" in m:
        mRNA = data.iloc[index, 2]
        recurrent.loc[len(recurrent)] = [mRNA]
    index = index + 1

  result = ttest_ind(nonrecurrent, recurrent)
  print(result)

  if result.statistic < 0 and result.pvalue < 0.05:
      tally = tally + 1

  print(tally)
```
